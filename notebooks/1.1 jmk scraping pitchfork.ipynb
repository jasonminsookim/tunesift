```python
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup
from requests import get
import time
import datetime

from src.func import func
```

```python
RP_START = "https://pitchfork.com/reviews/albums/?page=" # There should be a page number
AR_START = "https://pitchfork.com"

rp_page_number = 1
max_page_scraping = 5

# Create empty lists to append to
album_titles = []
album_artists = []
album_scores = []
album_genres = []
album_reviewers = []
album_review_texts = []
album_review_dates = []
```

```python
while(get(RP_START + str(rp_page_number)).status_code == 200):
    time.sleep(.5)
    
    response = get(RP_START + str(rp_page_number))
    soup = BeautifulSoup(response.content, 'html.parser')
    page_reviews = soup.find_all('div', class_ = 'review')
    
    for review in page_reviews:
        # Album title and artist
        album_title = review.find('div', class_ = 'review__title').find('h2', class_ = 'review__title-album').text
        album_artist = review.find('div', class_ = 'review__title').find('ul', class_ = 'artist-list review__title-artist').text
        
        # In depth album information from review link
        time.sleep(.5)
        album_review_link = review.find('a', class_ = 'review__link')['href']
        album_response = get(AR_START + album_review_link)
        album_soup = BeautifulSoup(album_response.content, 'html.parser')
        
        album_score = album_soup.find('div', class_ = 'score-circle').text
        try:
            album_genre = album_soup.find('li', class_ = 'genre-list__item').text
        except:
            album_genre = "NA"
        album_reviewer = album_soup.find('a', class_ = 'authors-detail__display-name').text
        album_review_text = album_soup.find('div', class_ = 'review-detail__article-content').text
        album_review_date = func.convert_to_date(album_soup.find('time', class_ = 'pub-date')['datetime'])
        
        # Append individual album elements into the lists 
        album_titles.append(album_title)
        album_artists.append(album_artist)
        album_scores.append(album_score)
        album_genres.append(album_genre)
        album_reviewers.append(album_reviewer)
        album_review_texts.append(album_review_text)
        album_review_dates.append(album_review_date)

    rp_page_number += 1
```

```python
pitchfork_df = pd.DataFrame({'title':album_titles,'artist':album_artists,'score':album_scores,'genre':album_genres,
                            'reviewer': album_reviewers, 'album_review_text': album_review_texts,
                             'album_review_date':album_review_dates})
pitchfork_df
```
